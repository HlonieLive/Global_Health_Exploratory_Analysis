# 3

```python
# Removing Unwanted columns

df.drop(columns = "PrimarySchoolEnrollmentMale", inplace=True)
df.drop(columns = "PrimarySchoolEnrollmentFemale", inplace=True)
df
```

```python
# Fill missing values

df.interpolate(method = "index", limit_direction = "both", inplace = True)
df
```

```python
# Checking duplicates

df.duplicated().sum()
```

```python
# Showing information to check everything such as data types of the column and etc.

df.info()
```

```python
# Renaming the GNI column to GrossNationalIncome
df.rename(columns = {"GNI": "GrossNationalIncome"}, inplace = True)
```

```python
# Checking the summary of the data set

df.describe()
```

```python
# Changing the data types of the numeric columns to Integers because using different data types might affect the accuracy of the analysis.

df["Under15"] = pd.to_numeric(df["Under15"]).astype(int)
df["Over60"] = pd.to_numeric(df["Over60"]).astype(int)
df["FertilityRate"] = pd.to_numeric(df["FertilityRate"]).astype(int)
df["ChildMortality"] = pd.to_numeric(df["ChildMortality"]).astype(int)
df["CellularSubscribers"] = pd.to_numeric(df["CellularSubscribers"]).astype(int)
df["LiteracyRate"] = pd.to_numeric(df["LiteracyRate"]).astype(int)
df["GrossNationalIncome"] = pd.to_numeric(df["GrossNationalIncome"]).astype(int)
df
```

```python
# Renaming the DataFrame

health_and_demogrphics_df = df
health_and_demogrphics_df
```

```python
# Verifying that there are no missing values remaining

health_and_demogrphics_df.isnull().sum()
```

# 9.

```python
# A box-and-wisker distribution for LifeExpectancy

A = health_and_demogrphics_df["LifeExpectancy"].to_list()
B = health_and_demogrphics_df["Region"].to_list()

sns.boxplot(x = A, y = B, data = health_and_demogrphics_df)

plt.title("The Distribution of Life Expentancy Across Regions", weight = "bold")
plt.xlabel("LifeExpectancy", weight = "bold")
plt.ylabel("Regions", weight = "bold")

plt.show()
```

With the observation from this box-and-whisker plot, I can safely say that the life expentancy distribution in Africa varies between the ages of 54-63 years, while it is between 65 and 77 years in most regions namely the Eastern Mediterranean, Americas, Western Pacific and South-East Asia. In Europe it varies between the ages of 75-78 years. This indicates that the majority of deaths in Africa occur the ages of 54-63 years, in Europe it is between 75-85 years, and in the remaining regions it is between 65-77 years. I can conclude that people from Africa do not live longer than people in Europe and other regions. People from Europe live the longest out of all regions.

```python
# Order the DataFrame by GrossNationalIncome

x = health_and_demogrphics_df.sort_values(by = ["GrossNationalIncome"], ascending = False)
x
```

```python
# Bar graph showing the distribution of LifeExpectancy related to Wealth across Regions

x = health_and_demogrphics_df.sort_values(by = ["GrossNationalIncome"], ascending = False)

A = x["Region"].to_list()
B = x["LifeExpectancy"].to_list()

plt.barh(A, B, color = "slategrey")
plt.title("Distribution of Life Expectancy related to Wealth across Regions", weight = "bold")
plt.xlabel("LifeExpectancy", weight = "bold")
plt.ylabel("Regions", weight = "bold")

plt.show()
```

According to this visualisation graph above, the wealtheir the region is, the more Life Expectance it has. Meaning that people in Eastern Mediterranean region are expected to live more than people who are in Africa and the opposite is correct because Africa has less Gross National Income compared to Eastern Mediterranean. However, there are some regions who don't follow this relation, such as South-East Asia which has less Gross National Income but more Life expentancy than Africa.

```python
# Bar graph showing the distribution of wealth related to Fertility Rate across Regions

y = health_and_demogrphics_df.sort_values(by = ["FertilityRate"], ascending = False)

C = y["Region"].to_list()
D = y["GrossNationalIncome"].to_list()

plt.barh(C, D, color = "lightblue")
plt.title("Distribution of Wealth related to Fertility Rate across Regions", weight = "bold")
plt.xlabel("GrossNationalIncome", weight = "bold")
plt.ylabel("Regions", weight = "bold")

plt.show()
```

```python
# Order by FertilityRate

y = health_and_demogrphics_df.sort_values(by = ["FertilityRate"], ascending = True)
y
```

According to this visualisation graph above, the wealtheir the region is, the less Fertility Rate it has. Meaning that Eastern Mediterranean region has less Fertility Rate than South-East Asia and the opposite is correct, so because it has more Gross National Income. However, there are some regions who don't follow this relation, such as Americas which has both Fertility Rate and Gross National Income less than those of Eastern Mediterranea.

# 11.

```python
# Order the DataFrame by Region

z = health_and_demogrphics_df.sort_values(by = ["Region"], ascending = True)
z
```

```python
# Count the number of regions

z.groupby(["Region"]).count()
```

```python
# Plot to represent the relationship between literacy and life expectancy by region

D = z["LifeExpectancy"].to_list()
E = z["LiteracyRate"].to_list()

plt.figure(figsize = (9, 6))
plt.scatter(E, D)

plt.title("Literacy Rate and Life Expectancy by Region", weight = "bold")
plt.xlabel("Life Expectation", weight = "bold")
plt.ylabel("Literacy Rate", weight = "bold")

plt.show()
```

```python
# The relationship between Literacy Rate and Life Expectancy by African Countries

F = z["LifeExpectancy"].head(46).to_list()
G = z["LiteracyRate"].head(46).to_list()

plt.figure(figsize = (9, 6))
plt.scatter(F, G)

plt.title("Literacy Rate and Life Expectancy by African Countries", weight = "bold")
plt.xlabel("Life Expectancy", weight = "bold")
plt.ylabel("Literacy Rate", weight = "bold")

plt.show()
```

The two plots indicate that Life Expectation and Literacy Rate have a positive relashionship in both the regional and African countries analysis. This means that, long Life Expectation is often associated with higher Literacy Rates.
